```python
from math import sin
import numpy as np
import matplotlib.pyplot as plt

xs = np.arange(0, 6, 6/360)
```

```python
def y(x, e):
  random_n = np.random.rand(*xs.shape)
  return (15 + sin(x) + random_n[e]*0.1)
```

```python
y(1, 0)
```

```python
ys = []
for x in xs:
  a = int(x/10)
  ys.append(y(x, a))

plt.plot(xs, ys, '.')
```

```python
s = 0
a = 15
n = 10
random_n = np.random.rand(*xs.shape)
for i in range(n):
  for x in xs :
    for y in ys :
      s += (a + sin(x) + random_n[int(x/10)]*0.1 - y)**2
print(s)
```

```python
def L(a, n) :
  s = 0
  random_n = np.random.rand(*xs.shape)
  for i in range(n):
    for x in xs :
      for y in ys :
        s += (a + sin(x) + random_n[int(x/10)]*0.1 - y)**2
  return s
```

```python
l = []
a_l = []
for a in range(20) :
  l.append(L(a, 10))
  a_l.append(a)
plt.plot(a_l, l)
print(l)
```
