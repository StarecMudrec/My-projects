```python
from math import sin, cos, atan2, sqrt
import matplotlib.pyplot as plt

V0 = 10
angle = 1
mv = 5
fv = 5000
M0 = 100
MF0 = 300
g = 10

dt = 0.01
t = 0
mf = MF0
m = M0 + mf
x = 0
y = 0
vx = V0 * cos(angle)
vy = V0 * sin(angle)

x_arr = []
y_arr = []
colors  = []

while y >= 0 :
  t += dt
  if mf > 0 :
    mf -= dt*fv
    m -= dt*mv
    vx += dt * (fv/m * cos(angle))
    vy += dt * (fv/m * sin(angle) - 10)
    colors.append("blue")
  else :
    colors.append("red")
    vy -= g*dt
  x += vx*dt
  y += vy*dt
  x_arr.append(x)
  y_arr.append(y)
  angle = atan2(vy, vx)

plt.scatter(x_arr, y_arr, color = colors)
```
