```python
import math as m
import matplotlib.pyplot as plt
#M = 1
L = 1
g = 10
A = 90 * m.pi / 180
y = L*(m.cos(A))
x = L*(m.sin(A))
old_x = x
old_y = y
v = 0
dt = 0.01
ax = -1*m.sin(A)*m.cos(A)*g
ay = 1*g*(1 - m.cos(A)**2)
vx = ax*dt
vy = ay*dt
coord_x = []
coord_y = []
for i in range(1000):
    A = m.atan2(x,y)
    ax = -1*m.sin(A)*m.cos(A)*g
    ay = 1*g*(1 - m.cos(A)**2)
    vx += ax*dt
    vy += ay*dt
    old_x = x
    old_y = y
    x += vx*dt
    y += vy*dt
    L = m.hypot(x,y)
    new_x = x / L
    new_y = y / L
    vx = (new_x - old_x)/dt
    vy = (new_y - old_y)/dt


    x = new_x
    y = new_y
    L = 1
    #print(A)
    coord_x.append(x)
    coord_y.append(L - y)
    #plt.plot(i, (vx**2 + vy**2)/2, '.')
    #plt.plot(i, g*(L - y), '.')
    plt.plot(i, (vx**2 + vy**2)/2 + g*(L - y), '.')
    plt.plot(i, L - y, '.')
'''plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.plot(coord_x, coord_y, '.')'''
    #plt.plot(i, (vx**2 + vy**2)/2, '.')
```
